```
import cv2
import numpy as np
import matplotlib.pyplot as plt
I = cv2.imread('fruits.png')
I = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)
I_flipVertical = cv2.flip(I, 0) 
I_flipHorizontal = cv2.flip(I, 1)
I_flipBoth = cv2.flip(I, -1) 
plt.figure()
plt.imshow(I)
plt.figure()
plt.imshow(I_flipVertical)
plt.figure()
plt.imshow(I_flipHorizontal)
plt.figure()
plt.imshow(I_flipBoth)
```
