```
import cv2
import numpy as np
import matplotlib.pyplot as plt
I = cv2.imread('fruits.png')
I = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)
rows,cols, ch = I.shape
theta = 45
M = cv2.getRotationMatrix2D((cols/2,rows/2),theta,1)
I_rot = cv2.warpAffine(I,M,(cols,rows))
plt.figure()
plt.imshow(I)
plt.figure()
plt.imshow(I_rot)
```
