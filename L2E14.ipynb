```
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.transform import warp, swirl
I = cv2.imread('cat.png')
I = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)
I_swirl = swirl(I, rotation=0, strength=20, radius=120)
plt.figure()
plt.imshow(I)
plt.figure()
plt.imshow(I_swirl)
```
