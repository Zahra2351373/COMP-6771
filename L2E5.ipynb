```
import cv2
import matplotlib.pyplot as plt
f1 = cv2.imread('boat.png')
f2 = cv2.imread('cameraman.png')
alpha = 0.3
g = cv2.addWeighted(f1, 1-alpha, f2, alpha, 0)
plt.figure()
plt.imshow(f1)
plt.figure()
plt.imshow(f2)
plt.figure()
plt.imshow(g)
```
