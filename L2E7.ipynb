```
import cv2
import numpy as np
import matplotlib.pyplot as plt
I = cv2.imread('fruits.png')
I = cv2.cvtColor(I, cv2.COLOR_BGR2RGB)
rows,cols, ch = I.shape
tx = 100; ty = 50
M = np.float32([[1,0,tx],[0,1,ty]])
I_trans = cv2.warpAffine(I,M,(cols,rows))
plt.figure()
plt.imshow(I)
plt.figure()
plt.imshow(I_trans)
```
